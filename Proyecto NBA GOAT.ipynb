```python
from nba_api.stats.endpoints import playercareerstats
import pandas as pd
```

#PARA NUESTRO ANÁLISIS DE LOS GOATS, AVERIGUAMOS SUS ID'S EN LA CSV PLAYERS

```python
GOATs_IDS = {
    "Michael Jordan": "893",
    "Kobe Bryant": "977",
    "LeBron James": "2544"
}
```

```python
##CREAMOS LA FUNCION DE STATS PARA ACCEDER A LOS DATOS DE LOS TRES JUGADORES Y CREAR UN DF CORRESPONDIENTE DE CADA UNO
```

```python
def GOATs_stats(player_name, player_id):
    carrera = playercareerstats.PlayerCareerStats(player_id=player_id)
    df = carrera.get_data_frames()[0]
    df["PLAYER_NAME"] = player_name
    return df
```

```python
jordan_df = GOATs_stats("Michael Jordan", GOATs_IDS["Michael Jordan"])
kobe_df = GOATs_stats("Kobe Bryant", GOATs_IDS["Kobe Bryant"])
lebron_df = GOATs_stats("LeBron James", GOATs_IDS["LeBron James"])
```

##ELIMINAMOS COLUMNAS QUE NO HARÁN PARTE DE NUESTRO ANÁLISIS

```python
#DEFINIMOS MÉTRICAS POR MINUTOS JUGADOS A LOS FINES DE TENER UN ANÁLISIS LIBRE DE SESGOS
```

```python
goat_career_clean["PTS_PER_MIN"] = goat_career_clean["PTS"] / goat_career_clean["MIN"]
goat_career_clean["REB_PER_MIN"] = goat_career_clean["REB"] / goat_career_clean["MIN"]
goat_career_clean["AST_PER_MIN"] = goat_career_clean["AST"] / goat_career_clean["MIN"]
goat_career_clean["STL_PER_MIN"] = goat_career_clean["STL"] / goat_career_clean["MIN"]
goat_career_clean["BLK_PER_MIN"] = goat_career_clean["BLK"] / goat_career_clean["MIN"]
```

#DEFINIMOS LA UTILIZACIÓN DE MÉTRICA "PER 36" UTILIZADA EN NBA PARA FINES ESTADÍSTICOS

```python
goat_career_clean["PTS_PER_36"] = goat_career_clean["PTS_PER_MIN"] * 36
goat_career_clean["REB_PER_36"] = goat_career_clean["REB_PER_MIN"] * 36
goat_career_clean["AST_PER_36"] = goat_career_clean["AST_PER_MIN"] * 36
goat_career_clean["STL_PER_36"] = goat_career_clean["STL_PER_MIN"] * 36
goat_career_clean["BLK_PER_36"] = goat_career_clean["BLK_PER_MIN"] * 36
```

```python
GOAT_METRICS = [
    "PTS_PER_MIN", "PTS_PER_36",
    "REB_PER_MIN", "REB_PER_36",
    "AST_PER_MIN", "AST_PER_36",
    "STL_PER_MIN",
    "BLK_PER_MIN",
    "FG_PCT",
    "FG3_PCT",
    "FT_PCT",
    "STL_PER_36",
    "BLK_PER_36"]
```

```python
GOAT_TEST = goat_career_clean[
    ["PLAYER_ID", "PLAYER_NAME", "SEASON_ID"] + GOAT_METRICS
]
```

```python
#AGREGAMOS POR JUGADOR PARA TENER UNA COMPARATIVA
```

```python
goat_by_player = (
    GOAT_TEST
    .groupby(["PLAYER_ID", "PLAYER_NAME"], as_index=False)
    .mean(numeric_only=True)
)
```

```python
goat_by_player[[
    "PLAYER_NAME",
    "PTS_PER_MIN",
    "PTS_PER_36",
    "REB_PER_36",
    "AST_PER_36",
    "FG_PCT",
    "FG3_PCT",
    "FT_PCT",
    "STL_PER_36",
    "BLK_PER_36"
]]
```

```python
#CREAREMOS UN INDICE DE MÉTRICAS PARA PONDERAR CADA UNA Y EVALUAR A CADA JUGADOR
```

```python
from sklearn.preprocessing import MinMaxScaler

metrics_for_index = [
    "PTS_PER_36",
    "REB_PER_36",
    "AST_PER_36",
    "STL_PER_36",
    "BLK_PER_36",
    "FG_PCT",
    "FG3_PCT",
    "FT_PCT"
]


```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

goat_scaled = goat_by_player.copy()
goat_scaled[metrics_for_index] = scaler.fit_transform(
    goat_scaled[metrics_for_index]
)
```

```python
#ASIGNAMOS PESO A LAS METRICAS
```

```python

weights = {
    "PTS_PER_36": 0.25,
    "REB_PER_36": 0.15,
    "AST_PER_36": 0.15,
    "STL_PER_36": 0.05,
    "BLK_PER_36": 0.05,
    "FG_PCT": 0.15,
    "FG3_PCT": 0.10,
    "FT_PCT": 0.10
}
```

```python
goat_scaled["GOAT_INDEX"] = sum(
    goat_scaled[m] * w for m, w in weights.items()
)

goat_final = goat_scaled[
    ["PLAYER_NAME", "GOAT_INDEX"]
].sort_values("GOAT_INDEX", ascending=False)

goat_final


```
